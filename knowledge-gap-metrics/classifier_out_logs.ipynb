```python
f = open("content_selection.out_machine_learning_output", "r")
all_result = []
for line in f:
    if line.startswith("This Wikipedia "):
        r = {}
        m = re.search(r"([a-z]+wiki)", line)
        r['wiki'] = m.group(0)
        m = re.search(r"(\d+)", line)
        r['total_article'] = int(m.group(0))
        continue
    if line.startswith("There were already "):
        m = re.search(r"(\d+)", line)
        r['ccc_groundtruth'] = int(m.group(0))
        m = re.search(r"(\d+\.\d*)", line)
        r['ccc_groundtruth_p'] = float(m.group(0))
        continue
    if line.startswith("This algorithm CLASSIFIED "):
        m = re.findall(r"(\d+)", line)
        r['test_article'] = int(m[2])
        r['predict_positive'] = int(m[0])
        m = re.search(r"(\d+\.\d*)", line)
        r['predict_positive_p'] = float(m.group(0))
        continue
    if line.startswith("With "):
        m = re.search(r"(\d+\.\d*)", line)
        r['ccc_final_p'] = float(m.group(0))
        all_result.append(r)
        continue
```

```python
df = pd.DataFrame.from_records(all_result)
```
